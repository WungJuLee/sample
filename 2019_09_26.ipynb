```r
flights
```

```r
flights %>% filter(month 1 %in% 4)
```

```r
flights %>% filter(month == 1 | month == 12)
```

```r
delays <- flights %>%
          group_by(dest) %>%
          summarise(count = n(),
                    dist = mean(distance,na.rm = T),
                    delay = mean(arr_delay,na.rm=T)) %>%
          filter(count > 20, dest != "HNL")
```

```r
delays
```

```r
ggplot(data=delays) + aes(x=dest,y=count)+geom_bar(stat='identity')
```
